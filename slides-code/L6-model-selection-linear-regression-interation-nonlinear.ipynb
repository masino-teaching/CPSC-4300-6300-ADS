```python
# example of y= 1.7*x^2 + epsilon showing linear relationship between y and x^2

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 1.7 * x**2 + np.random.normal(0, 1, 100)

#plt.scatter(x, y)
plt.scatter(x**2, y)
#plt.legend(['y = 1.7*x^2 + epsilon', 'y = 1.7*x^2 + epsilon'])
plt.xlabel('x^2')
plt.ylabel('y')
plt.show()
```
